```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules # Corrected module name to mlxtend
import warnings
warnings.filterwarnings('ignore')
```

```python
titanic = pd.read_csv("/content/Titanic.csv")
titanic.head()
```

```python
titanic['Class'].unique()
```

```python
titanic['Class'].value_counts()
```

```python
titanic['Gender'].value_counts()
```

```python
df=pd.get_dummies(titanic,dtype='int')
df.head()
```

```python
titanic.shape[0]
```

```python
frequent_itemsets = apriori(df, min_support=0.5, use_colnames=True)
frequent_itemsets
```

```python
frequent_itemsets = apriori(df, min_support=0.2, use_colnames=True)
frequent_itemsets
```

```python
frequent_itemsets = apriori(df, min_support=0.1, use_colnames=True)
frequent_itemsets
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1,num_itemsets=len(frequent_itemsets))
rules
```
