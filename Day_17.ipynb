```python
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
df =sns.load_dataset('tips')
df
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
```

```python
df['smoker']=lb.fit_transform(df['smoker'])
df['sex']=lb.fit_transform(df['sex'])
df['time']=lb.fit_transform(df['time'])
df['day']=lb.fit_transform(df['day'])
```

```python
df.dtypes
```

```python
df.corr()
```

```python
import matplotlib.pyplot as plt
sns.heatmap(np.abs(df.corr()), cmap='Purples')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
sns.heatmap(np.abs(df.corr()), cmap='Purples', annot=True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Calculate the absolute correlation matrix
corr_matrix = np.abs(df.corr())

# Filter the correlation matrix to keep values greater than 0.5
filtered_corr_matrix = corr_matrix[corr_matrix > 0.5]

# Plot the heatmap with the filtered correlation matrix
sns.heatmap(filtered_corr_matrix, cmap='Greens')
plt.show()
```
